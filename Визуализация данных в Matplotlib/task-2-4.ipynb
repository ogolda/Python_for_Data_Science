```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format = 'svg'

plt.style.use('fivethirtyeight')
```

```python
df = pd.read_csv('creditcard.csv')
df.head()
```

```python
#переменной Class 
se = df['Class'].value_counts()
```

```python
se.plot(kind='bar')
plt.show()
```

```python
se.plot(kind='bar', logy = True)
plt.show()
```

```python
v1 = df[df['Class'].isin([1])]
v1 = v1[['V1']]
```

```python
v0 = df[df['Class'].isin([0])]
v0 = v0[['V1']]
```

```python
plt.hist(v1, bins=20, color='grey', alpha = 0.5, edgecolor='black', density = True, orientation = "horizontal")
plt.hist(v0, bins=20, color='red', alpha = 0.5, edgecolor='black', density = True,orientation = "horizontal")
plt.plot()
plt.xlabel('V1')
plt.legend(labels=['Class 0', 'Class 1'])
```
